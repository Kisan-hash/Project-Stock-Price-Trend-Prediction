```python
# ============================================
# üìà Stock Price Trend Prediction
# Regression Model
# ============================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

print("Starting Stock Price Prediction Project üöÄ")

# -------- 1Ô∏è‚É£ Generate Synthetic Stock Data --------
np.random.seed(42)

days = 200
time = np.arange(days)

price = 100 + time * 0.5 + np.random.normal(0, 5, days)

data = pd.DataFrame({
    "Day": time,
    "Price": price
})

print("Dataset Created Successfully ‚úÖ")
print(data.head())

# -------- 2Ô∏è‚É£ Prepare Data --------
X = data[["Day"]]
y = data["Price"]

# -------- 3Ô∏è‚É£ Train Model --------
model = LinearRegression()
model.fit(X, y)

print("Model Trained Successfully ‚úÖ")

# -------- 4Ô∏è‚É£ Predict Future Prices --------
future_days = np.arange(days, days + 30).reshape(-1, 1)
future_predictions = model.predict(future_days)

# -------- 5Ô∏è‚É£ Plot Results --------
plt.figure()

plt.plot(data["Day"], data["Price"])
plt.plot(future_days, future_predictions)

plt.xlabel("Day")
plt.ylabel("Stock Price")
plt.title("Stock Price Prediction")
plt.legend(["Historical Data", "Predicted Future Trend"])

plt.show()

# -------- 6Ô∏è‚É£ Predict Specific Day --------
next_day = np.array([[days + 1]])
predicted_price = model.predict(next_day)

print("\nPredicted Price for Next Day:", round(predicted_price[0], 2))
print("\nProject Executed Successfully üöÄ")
```
